```python
def filter(data):
    result=[]
    
    
    beginprom1=pd.to_datetime(df_prom.iloc[0,1])
    endprom1=pd.to_datetime(df_prom.iloc[0,2])
    beginprom2=pd.to_datetime(df_prom.iloc[1,1])
    endprom2=pd.to_datetime(df_prom.iloc[1,2])
    beginprom3=pd.to_datetime(df_prom.iloc[2,1])
    endprom3=pd.to_datetime(df_prom.iloc[2,2])
    beginprom4=pd.to_datetime(df_prom.iloc[3,1])
    endprom4=pd.to_datetime(df_prom.iloc[3,2])
    beginprom5=pd.to_datetime(pd.to_datetime("9/1/2015"))
    endprom5=pd.to_datetime("6/9/2015")
    beginprom6=pd.to_datetime("11/20/2015")
    endprom6=pd.to_datetime("11/27/2015")
    
    
    for  i in data['Date']:
        if beginprom1<= i <= endprom1:
            result.append("Promo")
        elif beginprom2 <= i <= endprom2:
            result.append("Promo")
        elif beginprom3 <= i <= endprom3:
             result.append("Promo")
        elif beginprom4 <= i <= endprom4:
             result.append("Promo")
        elif beginprom5 <= i <= endprom5:
            result.append("Promo")
        elif beginprom6 <= i <= endprom6:
             result.append("Promo")
        else:
            result.append("Free")
   

    return result
```

Lets find the non-promotion periods and lets assign it into the variable named as df_nonProm

```python
df_nonProm=df1[df1['season']==0]
df_nonProm.shape
```

Now lets apply aggregation on a weekly basis

```python
df1_agg = df_nonProm[["StoreCode","ProductCode","SalesQuantity","Week_Number"]].groupby(['Week_Number',"StoreCode", "ProductCode"]).agg({ 'SalesQuantity':'sum'}).reset_index()
```

```python
df1_agg.shape
```

```python
df1_agg # This data frame gives the sales quantity of entire of WeekNum-StoreCode and ProductCode pair
```

Product with higher average weekly sale per store during non-promotion periods will be called â€œFast itemsâ€ and items with lower weekly average sale per store will be labeled as â€œSlow itemsâ€, items in between will be called â€œMedium itemsâ€. Grouping parameters selection is left to you. Apply similar approach to Stores as well.

Now we are only interested in ProductCodes and try to find average weekly sale per store. Below query returns average sales quantity of specific ProductCode per store in particular week. 

```python
df1_agg2 = df_nonProm[["StoreCode","ProductCode","SalesQuantity","Week_Number"]].groupby(['Week_Number', "ProductCode"],as_index='False').agg({'SalesQuantity':'mean'}).reset_index()
df1_agg2
```

Finally if we group by again it by week , this will give the average weekly sales of products per week.

```python
#Mean of all Products sales quantities (weekly average)
df1_agg2.groupby(['ProductCode'],as_index='False').agg({'SalesQuantity':'mean'}).reset_index()
```

Now let's visualize it. Remember that this results not among the promotion periods , this is why there is some gaps in the X-Axis


```python
trace=df1_agg2.groupby(['ProductCode'],as_index='False').agg({'SalesQuantity':'mean'}).reset_index()

trace1 = go.Bar(x=trace['ProductCode'], y=trace['SalesQuantity'], opacity=0.85, name="Mean Sales", marker=dict(color ='#002300'))
layout = go.Layout(title=go.layout.Title( text='Average Weekly Sale per Store During non-promotion Periods', xref='paper', x=0 ),
xaxis=go.layout.XAxis(
        title=go.layout.xaxis.Title(
            text='Product Code',
            font=dict(
                family='Courier New, monospace',
                size=18,
                color='#7f7f7f'
            )
        )
    ),
    yaxis=go.layout.YAxis(
        title=go.layout.yaxis.Title(
            text='Mean Sales',
            font=dict(
                family='Courier New, monospace',
                size=18,
                color='#737f7f'
            )
        )
    )
)
fig = go.Figure(data=[trace1], layout=layout);
iplot(fig)



```
